```python
from astropy.table import Table
import matplotlib.pyplot as plt

# read the m37 data csv file
t = Table.read("summary_m37_joined.csv", format="csv")

# define membership
members = t[t["PSS"] > 0.9]
nonmembers = t[t["PSS"] <= 0.9]

pmra = t["pmRA"]
pmdec = t["pmDE"]

# Plot
plt.figure(figsize=(6,6))

# Nonmembers - smaller and fainter
plt.scatter(nonmembers["pmRA"], nonmembers["pmDE"],
            s=5, alpha=0.2, color="gray", label="Non-members")

# Members - larger and darker
plt.scatter(members["pmRA"], members["pmDE"],
            s=12, alpha=0.8, color="blue", label="Members")

plt.xlabel("Proper motion in RA")
plt.ylabel("Proper motion in Dec")
plt.title("Proper motion in RA vs Proper motion in Dec")
plt.grid(alpha=0.3)
plt.legend()

plt.show()
```
