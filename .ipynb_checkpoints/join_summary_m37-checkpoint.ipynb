```python
import astropy.table as at

# Read both csv files
t1 = at.Table.read('summary_m37.csv', format='csv')
t2 = at.Table.read('summary_m37_xmatch_gaiadr3.csv', format='csv')

# Join the the two tables on the 'id' column
t_joined_id = at.join(t1, t2, keys='id')

# Save the joined table to a new file
t_joined_id.write('summary_m37_joined.csv', format='csv', overwrite=True)
```
