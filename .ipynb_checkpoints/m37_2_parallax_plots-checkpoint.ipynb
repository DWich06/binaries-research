```python
from astropy.table import Table
import matplotlib.pyplot as plt

# Load Gaia table
t = Table.read("summary_m37_joined.csv", format="csv")

# Extract columns
parallax = t["Plx"]
pmra = t["pmRA"]
pmdec = t["pmDE"]

# --- Parallax vs pmRA ---
plt.figure(figsize=(6,6))
plt.scatter(pmra, parallax, s=8, alpha=0.6)
plt.xlabel("pmRA")
plt.ylabel("Parallax")
plt.title("Parallax vs Proper Motion in RA")
plt.grid(alpha=0.3)
plt.show()

# --- Parallax vs pmDE ---
plt.figure(figsize=(6,6))
plt.scatter(pmdec, parallax, s=8, alpha=0.6)
plt.xlabel("pmDE")
plt.ylabel("Parallax")
plt.title("Parallax vs Proper Motion in Dec")
plt.grid(alpha=0.3)
plt.show()
```
