```python
from astropy.table import Table
import matplotlib.pyplot as plt

# read the m37 data csv file
t = Table.read("summary_m37_joined.csv", format="csv")

g = t["Gmag"]
bp_rp = t["BP-RP"]

# create the plot
plt.figure(figsize=(4, 5))
plt.scatter(bp_rp, g, s=5)
plt.gca().invert_yaxis()
plt.xlabel("BPâˆ’RP")
plt.ylabel("G")
plt.title("G vs BP-RP")
plt.grid(alpha=0.3)

plt.show()
```
