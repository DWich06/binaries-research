```python
import astropy.table as at
import matplotlib.pyplot as plt

# Read the m37 data csv file
t = at.Table.read('summary_m37_joined.csv', format='csv')

# Create the plot
plt.figure(figsize=(8, 6))

# Plot the stars
plt.scatter(t['bv'], t['v'], s=10, color='blue', alpha=1)

# Invert y axis
plt.gca().invert_yaxis()

# Style and formatting
plt.xlabel('V-B')
plt.ylabel('V')

plt.grid(True, alpha=0.3)

plt.show()
```
